```python
class FoodItem:
    def __init__(self, food_id, name, quantity, price, discount, stock):
        self.food_id = food_id
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock


class RestaurantMenu:
    def __init__(self):
        self.menu = []

    def add_food_item(self, name, quantity, price, discount, stock):3
        food_id = len(self.menu) + 1  
        food_item = FoodItem(food_id, name, quantity, price, discount, stock)
        self.menu.append(food_item)
        print(f"Added Food Item - FoodID: {food_id}, Name: {name}")

    def edit_food_item(self, food_id, name, quantity, price, discount, stock):
        for item in self.menu:
            if item.food_id == food_id:
                item.name = name
                item.quantity = quantity
                item.price = price
                item.discount = discount
                item.stock = stock
                print(f"Edited Food Item - FoodID: {food_id}, Name: {name}")
                return
        print(f"Food Item with FoodID {food_id} not found.")

    def view_all_food_items(self):
        print("List of All Food Items:")
        for item in self.menu:
            print(
                f"FoodID: {item.food_id}, Name: {item.name}, Quantity: {item.quantity}, Price: {item.price}, Discount: {item.discount}, Stock: {item.stock}"
            )

    def remove_food_item(self, food_id):
        for item in self.menu:
            if item.food_id == food_id:
                self.menu.remove(item)
                print(f"Removed Food Item - FoodID: {food_id}, Name: {item.name}")
                return
        print(f"Food Item with FoodID {food_id} not found.")


# Create a menu instance
menu = RestaurantMenu()

menu.add_food_item("Burger", "1 piece", 5.99, 0.2, 50)
menu.add_food_item("Pizza", "1 slice", 2.99, 0.15, 20)
menu.view_all_food_items()
menu.edit_food_item(1, "Big Burger", "1 piece", 6.99, 0.25, 40)
menu.remove_food_item(2)
menu.view_all_food_items()
```

```python
class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.orders = []

    def place_order(self, items):
        order = []
        for item_number in items:
            if item_number >= 0 and item_number < len(menu.food_items):
                selected_item = menu.food_items[item_number]
                order.append(selected_item)
            else:
                print(f"Invalid item number: {item_number}")
        if order:
            self.orders.append(order)
            print("Order placed successfully!")
        else:
            print("No items in the order.")

    def view_order_history(self):
        print("Order History:")
        for i, order in enumerate(self.orders, 1):
            print(f"Order {i}:")
            for item in order:
                print(
                    f"Name: {item['name']}, Quantity: {item['quantity']}, Price: {item['price']}"
                )

    def update_profile(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        print("Profile updated successfully!")


class Menu:
    def __init__(self):
        self.food_items = []

    def add_food_item(self, name, quantity, price):
        food_item = {"name": name, "quantity": quantity, "price": price}
        self.food_items.append(food_item)

    def display_menu(self):
        print("Menu:")
        for i, item in enumerate(self.food_items, 1):
            print(
                f"{i}. Name: {item['name']}, Quantity: {item['quantity']}, Price: {item['price']}"
            )


# Create a menu instance and add some food items
menu = Menu()
menu.add_food_item("Tandoori Chicken", "4 pieces", 240)
menu.add_food_item("Vegan Burger", "1 piece", 320)
menu.add_food_item("Truffle Cake", "500gm", 900)

# Main program
user = None

while True:
    print("\n--- Welcome to the Restaurant ---")
    print("1. Register")
    print("2. Login")
    print("3. Exit")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        full_name = input("Enter Full Name: ")
        phone_number = input("Enter Phone Number: ")
        email = input("Enter Email: ")
        address = input("Enter Address: ")
        password = input("Enter Password: ")

        user = User(full_name, phone_number, email, address, password)
        print("Registration successful!")

    elif choice == 2:
        if user is None:
            print("Please register or login first.")
        else:
            print(f"Welcome, {user.full_name}!")
            while True:
                print("\n--- User Menu ---")
                print("1. Place New Order")
                print("2. Order History")
                print("3. Update Profile")
                print("4. Logout")
                user_choice = int(input("Enter your choice: "))

                if user_choice == 1:
                    menu.display_menu()
                    selected_items = input(
                        "Enter the item numbers you want to order (separated by spaces): "
                    )
                    selected_items = [int(item) for item in selected_items.split()]
                    user.place_order(selected_items)

                elif user_choice == 2:
                    user.view_order_history()

                elif user_choice == 3:
                    full_name = input("Enter Full Name: ")
                    phone_number = input("Enter Phone Number: ")
                    email = input("Enter Email: ")
                    address = input("Enter Address: ")
                    password = input("Enter Password: ")
                    user.update_profile(
                        full_name, phone_number, email, address, password
                    )

                elif user_choice == 4:
                    print("Logging out...")
                    break

    elif choice == 3:
        print("Goodbye!")
        break
```
